```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ingreso
fx = lambda x: np.sqrt(x)* np.sin(x) # fx es una funcion lambda que depende de x
a= 1 # intervalo
b = 3 # intervalo
tramos = 4
```

```python
muestras = tramos + 1
xi = np.linspace(a,b, muestras)
print(xi)
# xi es la distribucion de los puntos entre a y b con la cantidad de muestras
fi = fx(xi)
```

```python
#salida
muestraslinea = muestras * 10 # el * 10 es para lograr la linea
xk = np.linspace(a,b, muestraslinea)
fk = fx(xk)
plt.plot(xi,fi, 'ro') # Marcados por un punto ROJO "ro"
plt.plot(xk,fk) # linea azul
plt.show()
```

```python
muestraslinea = muestras * 10
xk = np.linspace(a,b, muestraslinea)
fk = fx(xk)
plt.plot(xi,fi, 'ro')
plt.plot(xk,fk)
# trapecios
plt.fill_between(xi,0, fi, color='g') #relleno entre la funcion que pasa por los puntos xi, 0, fi, verde
for i in range(0,muestras,1):
    plt.axvline(xi[i], color='w') # linea vertical en cada punto x[i] en blanco
plt.show()
```

```python
# Calculo de areas
h = (b-a)/tramos #constante
suma = 0
xi = a
for i in range(0, tramos, 1): # recorrer cada uno de los tramos entre 0 y tramos en pasos de 1
    Atrapecio = h * (fx(xi)+fx(xi+h))/2 # A trapecio 
    suma += Atrapecio # se acumula en suma cada area
    xi += h # xi le sumamos h
integral = suma
print(f"Tramos = {tramos} Integral = {integral}")
```

```python
def trapecio(tramos):
    # ingreso
    fx = lambda x: np.sqrt(x)* np.sin(x) # fx es una funcion lambda que depende de x
    a= 1 # intervalo
    b = 3 # intervalo
    muestras = tramos + 1
    xi = np.linspace(a,b, muestras)
    # xi es la distribucion de los puntos entre a y b con la cantidad de muestras
    fi = fx(xi)
    #salida
    muestraslinea = muestras * 10
    xk = np.linspace(a,b, muestraslinea)
    fk = fx(xk)
    plt.plot(xi,fi, 'ro')
    plt.plot(xk,fk)
    # trapecios
    plt.fill_between(xi,0, fi, color='g')
    for i in range(0,muestras,1):
        plt.axvline(xi[i], color='w')
    plt.show()
    # Calculo de areas
    h = (b-a)/tramos #constante
    suma = 0
    xi = a
    for i in range(0, tramos, 1):
        Atrapecio = h * (fx(xi)+fx(xi+h))/2
        suma += Atrapecio
        xi += h
    integral = suma
    print(f"Tramos = {tramos} Integral = {integral}")
```

```python
def trapeciofx(tramos, fx, a, b):
    muestras = tramos + 1
    xi = np.linspace(a,b, muestras)
    # xi es la distribucion de los puntos entre a y b con la cantidad de muestras
    fi = fx(xi)
    #salida
    muestraslinea = muestras * 10
    xk = np.linspace(a,b, muestraslinea)
    fk = fx(xk)
    plt.plot(xi,fi, 'ro')
    plt.plot(xk,fk)
    # trapecios
    plt.fill_between(xi,0, fi, color='g')
    for i in range(0,muestras,1):
        plt.axvline(xi[i], color='w')
    plt.show()
    # Calculo de areas
    h = (b-a)/tramos #constante
    suma = 0
    xi = a
    for i in range(0, tramos, 1):
        Atrapecio = h * (fx(xi)+fx(xi+h))/2
        suma += Atrapecio
        xi += h
    integral = suma
    print(f"Tramos = {tramos} Integral = {integral}")
```
